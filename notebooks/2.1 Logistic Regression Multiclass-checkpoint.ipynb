## Import Libraries
The following libraries are imported for data handling, visualization, and building the machine learning model.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Iris Dataset
The Iris dataset is loaded from sklearn. It contains measurements of iris flowers used for multiclass classification.

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.keys() 
```

## Import Train-Test Split
The train_test_split function is imported to divide the dataset into training and testing subsets.

```python
from sklearn.model_selection import train_test_split
```

```python
X= iris.data
y= iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

## Import Logistic Regression
Logistic Regression is imported to perform multiclass classification on the Iris dataset.

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

## Train Logistic Regression Model
A Logistic Regression model is created and trained using the training dataset to learn the relationship between the features and the flower species.

```python
model.fit(X_train,y_train)
```

## Make Predictions and Evaluate Model
The trained model is used to make predictions on the test dataset. Model performance is evaluated using accuracy and a confusion matrix, which is visualized using a heatmap.

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
```
